#### e. ta reda p√• statistik f√∂r √•ldern: medelv√§rde, median, min, max, standardavvikelse,


```python
df_age = df_events[["Age"]].dropna(inplace=False).sort_values(by=["Age"])

median = df_age.median()
mean = df_age.mean()
mode = df_age.mode().squeeze()
min_age = df_age.iloc[[0]].squeeze() # Squeeze converts dataframe to a series
max_age = df_age.iloc[[-1]].squeeze()
print(f"The median age is {median.to_string()}")
print(f"The mean age is {mean.to_string()}")
print(f"The mode age is {mode}")
print(f"The minimum age is {min_age}")
print(f"The max age is {max_age}")
```
